```python
import requests

# Paste your GitHub Personal Access Token
TOKEN = "ghp_XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"

HEADERS = {
    "Authorization": f"token {TOKEN}",
    "Accept": "application/vnd.github.v3+json"
}


all_repos = []
page = 1

print(" Fetching repositories from GitHub...\n")

while True:
    url = f"https://api.github.com/user/repos?per_page=100&page={page}"
    response = requests.get(url, headers=HEADERS)

    if response.status_code != 200:
        print(" Failed to fetch repos. Check your token.")
        break

    data = response.json()
    if not data:
        break

    all_repos.extend(data)
    page += 1

print(f"Total repos found: {len(all_repos)}")

forked_repos = [r for r in all_repos if r["fork"]]

if not forked_repos:
    print(" No forked repositories found. Nothing to delete.")
    exit()

print(f" Forked repos found: {len(forked_repos)}\n")

for i, repo in enumerate(forked_repos, start=1):
    print(f"{i}. {repo['full_name']}")


choice = input(
    "\nEnter repo numbers (e.g. 1,3,5) OR type ALL to delete all forked repos: "
).strip().upper()


if choice == "ALL":
    selected_repos = forked_repos.copy()

else:
    try:
        selected_indexes = [int(x.strip()) for x in choice.split(",")]
    except ValueError:
        print(" Invalid input. Use numbers or ALL.")
        exit()

    selected_repos = []
    for i in selected_indexes:
        if 1 <= i <= len(forked_repos):
            selected_repos.append(forked_repos[i - 1])


if not selected_repos:
    print(" No valid repositories selected.")
    exit()


print("\n You selected the following repositories:\n")
for r in selected_repos:
    print(" -", r["full_name"])

confirm = input(
    "\nType DELETE to permanently delete ALL selected repos: "
)

if confirm != "DELETE":
    print(" Cancelled. Nothing was deleted.")
    exit()

print("\n Deleting repositories...\n")

for repo in selected_repos:
    delete_url = f"https://api.github.com/repos/{repo['full_name']}"
    resp = requests.delete(delete_url, headers=HEADERS)

    if resp.status_code == 204:
        print(f"Deleted: {repo['full_name']}")
    else:
        print(
            f" Failed: {repo['full_name']} | "
            f"Status: {resp.status_code}"
        )

print("\n Done. Selected repositories processed.")
```
